```python
import json
import spacy
import os
```

```python
nlp = spacy.load('de_core_news_md')

corpath = 'corpus_selenium'
for file_name in os.listdir(corpath):
    print(file_name)
    with open(rf'{corpath}/{file_name}', 'r') as file:
        data = json.load(file)

    # Access the poem lines
    lines = data['poem']

    for line_num, line_data in lines.items():
        text = line_data['text']

        # Tokenize the line using spaCy
        doc = nlp(text)
        tokens = [token.text for token in doc]

        # Add the tokens to the "line.X" key
        line_data['tokens'] = tokens

    # Reopen the file in write mode and write the updated JSON data
    with open(rf'{corpath}/{file_name}', 'w') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)
```
